```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp("Tesla Inc is going ot acquire Twitter for $45 billion")

for ent in doc.ents:
    print(ent.text,"|",ent.label_,"|",spacy.explain(ent.label_))
```

```python
from spacy import displacy

displacy.render(doc, style = 'ent')
```
